# Task 2 — Teacher Forcing & Exposure Bias

### Training Autoregressive Decoders
When training a Seq2Seq decoder, the model generates words step-by-step. If the model makes a mistake early in the sentence (e.g. predicting "Le" instead of "La"), this mistake cascades, causing all subsequent predictions to degrade. This makes training extremely slow to converge.

### Teacher Forcing vs. Free-Running
To solve this, we use **Teacher Forcing**:
- **With Teacher Forcing**: During training, at each step, instead of feeding the decoder's own prediction as the next input, we feed the **ground-truth** target token from the training corpus. This keeps the model on the correct path and accelerates training convergence.
- **Without Teacher Forcing (Free-Running)**: The decoder must rely entirely on its own previous step prediction, representing the exact behavior needed during testing (inference).

### Exposure Bias
While Teacher Forcing improves training, it creates **Exposure Bias** — a mismatch between training and testing. Since the decoder is exposed to perfect inputs during training, it becomes fragile at inference time when a single early mistake causes the sequence translation to fall apart.
