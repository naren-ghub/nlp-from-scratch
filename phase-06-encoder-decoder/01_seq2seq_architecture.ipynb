# Task 1 — Seq2Seq Mapping & The Encoder-Decoder Paradigm

### Sequence-to-Sequence Mapping
Standard sequence models (like text classifiers) map sequences to a single output classification (Many-to-One), or translate word-by-word with identical lengths (Many-to-Many Synced). However, tasks like **Machine Translation** or **Text Summarization** require mapping an input sequence of length $N$ to an output sequence of a *different* length $M$ (Many-to-Many Un-synced).

### The Encoder-Decoder Stack
To map variable lengths, we use the **Seq2Seq Encoder-Decoder** framework:
1. **Encoder (RNN)**: Processes the source sequence step-by-step and compresses it into a single vector (the final hidden state, or **Context Vector**).
2. **Decoder (RNN)**: Unfolds this context vector to generate the target sequence word-by-word autoregressively.



Special tokens (`<PAD>`, `<SOS>`, `<EOS>`, `<UNK>`) are used to standardize input lengths and signal translation start and end steps.
