# Task 1 — POS Tagging & Dependency Parsing

### Introduction
Part-of-Speech (POS) Tagging and Dependency Parsing reveal the grammatical structure of a sentence.

1. **POS Tagging**: Labels words as Noun, Verb, Adjective, etc.
2. **Dependency Parsing**: Captures binary syntactic relationships between words, constructing a parse tree showing how modifier words link to head words (usually root verbs).

### Core Use Cases
- **Grammar Correction**: Analyzing sentences to spot incorrect parts-of-speech or agreement errors.
- **Information Extraction**: Identifying actions (verbs) and actors (nouns) to map knowledge graphs.
- **Question Answering**: Finding the syntactic focus of a user's question to pull answers.



### Step 1 — Tag POS and Dependencies using spaCy


```python
import spacy
from spacy import displacy
import pandas as pd

# Load spaCy model
try:
    nlp = spacy.load("en_core_web_sm")
except OSError:
    from spacy.cli import download
    download("en_core_web_sm")
    nlp = spacy.load("en_core_web_sm")

sentence = "The smart parser analyzed the complex sentence structure."
doc = nlp(sentence)

# Format tokens
tokens_data = []
for token in doc:
    tokens_data.append({
        'Token': token.text,
        'POS': token.pos_,
        'Tag': token.tag_,
        'Dependency': token.dep_,
        'Head': token.head.text,
        'Explanation': spacy.explain(token.pos_)
    })

print(pd.DataFrame(tokens_data))

```

### Step 2 — Visualize the Dependency Parse Tree


```python
# Visualize the syntactic parse tree
displacy.render(doc, style="dep", jupyter=True, options={"distance": 110})

```

### Step 3 — User-Defined Interactive Evaluation Function

```python
def evaluate_pos_tagging(text):
    """
    User-defined evaluation function. Parses text using spaCy and displays the POS dependency table.
    """
    custom_doc = nlp(text)
    data = []
    for token in custom_doc:
        data.append({
            'Token': token.text,
            'POS': token.pos_,
            'Dependency': token.dep_,
            'Head': token.head.text
        })
    print("\nParsed POS Tagging & Dependencies:")
    display(pd.DataFrame(data))
    
    print("\nSyntactic Dependency Visual:")
    displacy.render(custom_doc, style="dep", jupyter=True, options={"distance": 110})

# Evaluate with custom input
evaluate_pos_tagging("Natural Language Processing changes how computers read text.")

```
