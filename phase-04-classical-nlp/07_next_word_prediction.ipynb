# Task 7 — Next Word Prediction

### Introduction
Next Word Prediction is the task of predicting the most probable word to follow a given context.

We implement an **N-gram Language Model** (specifically a Trigram model) trained on NLTK text. The model estimates probabilities of the next word $w_3$ based on the preceding bigram context $(w_1, w_2)$:
$$P(w_3 | w_1, w_2) = \frac{\text{Count}(w_1, w_2, w_3)}{\text{Count}(w_1, w_2)}$$

### Core Use Cases
- **Smart Keyboards**: Autocomplete suggestions on mobile screens as a user types.
- **Search Suggestions**: Completing search strings in real-time as users enter letters.
- **Language Generation**: Providing the probability foundation for sequence text generation.



### Step 1 — Build Trigram Probabilities from a Corpus


```python
import nltk
from nltk.tokenize import word_tokenize
from collections import defaultdict, Counter

nltk.download('punkt', quiet=True)
nltk.download('webtext', quiet=True)
from nltk.corpus import webtext

# Load sample text corpus
corpus_words = [w.lower() for w in webtext.words() if w.isalpha()]
print(f"Corpus size: {len(corpus_words)} words")

# Build bigrams and trigrams
trigrams = list(nltk.trigrams(corpus_words))

# Transitions dictionary: context (w1, w2) -> next word candidate counts
transitions = defaultdict(Counter)
for w1, w2, w3 in trigrams:
    transitions[(w1, w2)][w3] += 1

def predict_next_word(w1, w2):
    context = (w1.lower(), w2.lower())
    candidates = transitions[context]
    if not candidates:
        return None
    # Return candidate word with highest occurrence
    return candidates.most_common(1)[0][0]

# Sample prediction
print("Test Context: ('going', 'to') -> Prediction:", predict_next_word("going", "to"))

```

### Step 2 — User-Defined Interactive Evaluation Function
Enter two words as a prompt (e.g. `'going to'` or `'i want'`) to evaluate the model's prediction of the next word.


```python
def evaluate_next_word_prediction(prompt):
    """
    User-defined evaluation function. Takes a prompt of words, extracts the last two, and predicts the next.
    """
    words = [w.lower() for w in word_tokenize(prompt) if w.isalpha()]
    if len(words) < 2:
        print("Error: Please enter a prompt containing at least 2 words.")
        return
        
    w1, w2 = words[-2], words[-1]
    prediction = predict_next_word(w1, w2)
    
    print(f"Prompt Context: '... {w1} {w2}'")
    if prediction:
        print(f"Predicted Next Word: '{prediction}'")
        # List other possibilities
        others = transitions[(w1, w2)].most_common(5)
        print("Top candidates and counts:")
        for candidate, count in others:
            print(f" - {candidate}: {count} times")
    else:
        print("No prediction found for this word combination in the corpus.")

# Test custom inputs
evaluate_next_word_prediction("going to")
evaluate_next_word_prediction("i want")

```
