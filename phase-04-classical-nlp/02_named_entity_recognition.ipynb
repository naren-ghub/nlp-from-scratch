# Task 2 — Named Entity Recognition (NER)

### Introduction
Named Entity Recognition (NER) identifies proper nouns in text and classifies them into categories (e.g. Person, Organization, Location, Date).

We will use **spaCy**'s pre-trained pipeline (`en_core_web_sm`) to run NER, visualize entities in context, and analyze occurrences.

### Core Use Cases
- **Content Tagging**: Automatically tagging people, places, and brands in blog posts or news articles.
- **Search Recommendation**: Parsing entity searches to direct queries to relevant sections.
- **Customer Support**: Extracting order IDs, locations, and names from email tickets.



### Step 1 — Load spaCy and Extract Entities


```python
import spacy
from spacy import displacy
import pandas as pd

# Download/Load spaCy model
try:
    nlp = spacy.load("en_core_web_sm")
except OSError:
    from spacy.cli import download
    download("en_core_web_sm")
    nlp = spacy.load("en_core_web_sm")

news_article = (
    "Google CEO Sundar Pichai announced new AI research centers in London, Paris, and Bengaluru. "
    "The initiative, launching in December 2026, represents a $500 million investment. "
    "Key researchers from MIT and Stanford University will join the centers."
)

doc = nlp(news_article)

# Extract entities
ents_data = []
for ent in doc.ents:
    ents_data.append({
        'Text': ent.text,
        'Label': ent.label_,
        'Start': ent.start_char,
        'End': ent.end_char,
        'Explanation': spacy.explain(ent.label_)
    })

print(pd.DataFrame(ents_data))

```

### Step 2 — Visualize Entities using displaCy


```python
# Visualize inline in Jupyter
displacy.render(doc, style="ent", jupyter=True)

```

### Step 3 — User-Defined Interactive Evaluation Function
Enter your own paragraph to view color-coded entities highlighted in context.


```python
def evaluate_named_entities(text):
    """
    User-defined evaluation function. Extracts named entities and renders them using displaCy.
    """
    custom_doc = nlp(text)
    if not custom_doc.ents:
        print("No named entities detected in the text.")
        return
        
    print("Entities Found:")
    for ent in custom_doc.ents:
        print(f" - {ent.text} ({ent.label_})")
        
    print("\nEntity Context Visualization:")
    displacy.render(custom_doc, style="ent", jupyter=True)

# Test with custom input
evaluate_named_entities("Apple Inc. announced plans to build a new data center in Munich, Germany, by January 2027.")

```
