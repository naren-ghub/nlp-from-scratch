# Task 4 — Lexicon-Based Sentiment Analysis

### Introduction
Sentiment Analysis is the process of mining text to identify positive, negative, or neutral sentiment.

**Lexicon-based approaches** utilize a pre-scored dictionary of words (lexicon).
One of the most famous rule-based lexicons is **VADER** (Valence Aware Dictionary and sEntiment Reasoner). VADER uses rules (punctuation boosts, capitalizations, contractions, and negation markers) to calculate scores.

### Core Use Cases
- **Social Listening**: Monitoring live feeds (like Twitter/X) for rapid sentiment tracking.
- **Unlabeled Feedback Processing**: Gaining sentiment insights on user comments without waiting for training labels.
- **Fast Prototyping**: Setting up baseline sentiment systems in production quickly.



### Step 1 — Load VADER and inspect token intensity scores


```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd

nltk.download('vader_lexicon', quiet=True)
analyzer = SentimentIntensityAnalyzer()

# Verify token intensities
tokens = ['good', 'bad', 'excellent', 'terrible', 'not bad', 'not good']
results = []
for tok in tokens:
    score = analyzer.polarity_scores(tok)
    results.append({
        'Token': tok,
        'Compound': score['compound'],
        'Positive': score['pos'],
        'Negative': score['neg'],
        'Neutral': score['neu']
    })

print(pd.DataFrame(results))

```

### Step 2 — User-Defined Interactive Evaluation Function
Enter your own sentence to calculate its positive, negative, neutral, and compound scores instantly.


```python
def evaluate_sentiment_lexicon(text):
    """
    User-defined evaluation function. Evaluates custom text sentiment using VADER.
    """
    scores = analyzer.polarity_scores(text)
    compound = scores['compound']
    
    if compound >= 0.05:
        verdict = "POSITIVE"
    elif compound <= -0.05:
        verdict = "NEGATIVE"
    else:
        verdict = "NEUTRAL"
        
    print(f"Sentence: '{text}'")
    print(f"Sentiment Verdict: {verdict}")
    print(f"Scores -> Compound: {compound:.4f} | Pos: {scores['pos']:.2f} | Neg: {scores['neg']:.2f} | Neu: {scores['neu']:.2f}\n")

# Test custom inputs
evaluate_sentiment_lexicon("The new updates are absolutely AMAZING!!!")
evaluate_sentiment_lexicon("Honestly, the product was not bad at all.")
evaluate_sentiment_lexicon("I hate waiting in long queue lists.")

```
