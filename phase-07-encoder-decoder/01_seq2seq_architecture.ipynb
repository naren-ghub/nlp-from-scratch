# Task 1 — Sequence-to-Sequence & Encoder-Decoder Architecture

## 1. What problem does this solve?

In previous phases, we explored architectures like RNNs and LSTMs. While powerful, they are constrained by a key limitation: they expect the **input sequence length to match the output sequence length** (or follow a direct step-by-step mapping). 

However, in many real-world NLP tasks, the input and output lengths vary dynamically. For example:
- **Machine Translation**: *"Je suis fatigué"* (3 words) $ightarrow$ *"I am tired"* (3 words), but *"Thank you"* (2 words) $ightarrow$ *"Ich danke Ihnen"* (3 words).
- **Text Summarization**: A 500-word article compressed into a 20-word summary.
- **Chatbots/Q&A**: A short question producing a detailed response.

To solve this variable-length sequence mapping problem, Cho et al. (2014) and Sutskever et al. (2014) proposed the **Encoder-Decoder (Seq2Seq)** framework.

---

## 2. Conceptual Architecture

The Encoder-Decoder architecture separates the model into two distinct recurrent neural networks connected by a single bottleneck state:

![Encoder-Decoder Seq2Seq Architecture](outputs/encoder-decoder.png)

### 1. The Encoder
- The Encoder takes the input sequence word-by-word (e.g., *"Je"*, *"suis"*, *"fatigué"*).
- It maintains a recurrent hidden state that updates at each time step.
- At the final word, the Encoder compresses the entire semantic meaning of the input sequence into its final hidden state vector. This final vector is called the **Context Vector** (or the semantic bottleneck).

### 2. The Context Vector
- The Context Vector acts as the final "summary" or translation pivot of the input sentence.
- It is a single, fixed-size vector containing the compressed representation of the source text.
- It is passed directly to the Decoder as its initial hidden state.

### 3. The Decoder
- The Decoder is initialized with the Context Vector.
- It generates the target sequence token-by-token.
- At each step, the Decoder uses its own previous output word and updated hidden state to predict the next word in the target language.

---

## 3. Special Tokens in Seq2Seq

To handle variable lengths and control generation logic, Seq2Seq models use three critical special tokens:

*   **`<PAD>` (Padding)**: Standardizes batch shapes. Since sentences in a batch have different lengths, shorter sentences are padded with `<PAD>` to match the maximum sequence length.
*   **`<SOS>` (Start of Sentence)**: Fed to the decoder at the very first step to kick off sequence generation.
*   **`<EOS>` (End of Sentence)**: Appended to target sentences. When the decoder predicts `<EOS>`, it signals that generation is complete, allowing the generation loop to terminate.
