# Task 1 — The Bottleneck and the Birth of Attention

### The Information Bottleneck Problem (Recap from Phase 6)
In traditional Seq2Seq architectures, the entire input sentence is compressed into a single fixed-size vector (called the **Context Vector** or **Bottleneck**). This design worked well for short sentences but failed drastically for longer sentences.

Imagine compressing an entire paragraph of 50 words into a single 128-dimensional vector, and then expecting the decoder to perfectly reconstruct it word-by-word. Information decay was inevitable because the network had to squeeze both semantic meanings, syntax structures, and lexical details into a single point.



---

### The Open-Book Analogy
The **Attention Mechanism** (introduced by Bahdanau et al. in 2014) directly addresses the information bottleneck by replacing the "closed-book" memory approach with an "open-book" retrieval approach.

Instead of forcing the network to memorize the entire source sentence before translating, the Attention Mechanism allows the Decoder to look back at the original source hidden states at *every single translation step*.



Under this analogy:
- **Closed-Book Exam**: The Decoder only has access to the final context vector. It must memorize everything and reproduce it.
- **Open-Book Exam**: The Decoder keeps the textbook open (access to all Encoder states $h_1, h_2, \dots, h_T$). When producing the next word, it turns to the most relevant page.

---

### Additive vs. Multiplicative Attention
There are two primary variants of sequence-to-sequence attention:
1. **Additive Attention (Bahdanau Attention)**:
   - Computes attention scores using a feed-forward network with a single hidden layer:
     $$score(s_{t-1}, h_i) = v_a^T \tanh(W_a s_{t-1} + U_a h_i)$$
   - Highly flexible, but computationally slower due to parameter matrices and activation steps.
2. **Multiplicative Attention (Luong Attention)**:
   - Simplifies the score computation using dot products or matrix multiplication:
     $$score(s_t, h_i) = s_t^T W_a h_i$$
   - Faster and more parameter-efficient, utilizing optimized matrix multiplication operations.

---

### Attention Alignment Matrix
By plotting attention weights across time steps, we obtain an **Alignment Matrix** (heatmap). This matrix illustrates exactly which source words the decoder focused on while generating each target word.

For example, when translating French adjectives that appear *after* nouns back into English (where adjectives appear *before* nouns), the attention weights shift accordingly, revealing the syntax alignment mapping learned by the model.
