# Task 6 — How Transformers Changed AI

### The Three Architecture Variants
Depending on how self-attention is stacked, Transformers are split into three main families:



1. **Encoder-only (Autoencoding - e.g., BERT)**:
   - Uses full bi-directional self-attention to see the entire context.
   - **Usecases**: Text Classification, Sentiment, Named Entity Recognition (NER), Feature Extraction.
2. **Decoder-only (Autoregressive - e.g., GPT family, Llama)**:
   - Uses masked self-attention to predict the next token given past tokens.
   - **Usecases**: Text Completion, Generative Writing, Code Generation, Chatbots.
3. **Encoder-Decoder (Seq2Seq - e.g., T5, BART)**:
   - Connects an encoder to a decoder utilizing cross-attention.
   - **Usecases**: Machine Translation, Summarization, Question Answering.

---

### The Transfer Learning Paradigm Shift
Before Transformers, NLP models had to be trained from scratch for every specific task, requiring large labeled datasets.

The parallelization of self-attention allowed models to be pre-trained on **massive, unlabeled web-scale text corpora**. The model learns grammar rules, world facts, and language patterns via unsupervised tasks (like Masked Language Modeling).



This pre-trained "base model" is then **fine-tuned** on a very small labeled dataset for a downstream task (e.g., classifying support tickets), saving massive computation costs.
