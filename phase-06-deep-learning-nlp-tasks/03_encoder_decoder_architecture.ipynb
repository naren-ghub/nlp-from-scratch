# Task 3 — Encoder-Decoder Architecture

Standard sequence-to-sequence mappings (like translating language or writing summaries) pose a major technical challenge: the **input sequence length does not match the output sequence length**. For instance:
- Input (French): *"Le chat est sur la table"* (6 tokens)
- Output (English): *"The cat is on the table"* (6 tokens, but word-for-word translation order could fail for other sentences, and sizes vary).
- Input: *"Thank you"* (2 words) $ightarrow$ Output (German): *"Vielen Dank"* (2 words) or *"Ich danke Ihnen"* (3 words).

To solve this, Cho et al. (2014) and Sutskever et al. (2014) introduced the **Encoder-Decoder (Seq2Seq)** framework.

### Conceptual Flow of Encoder-Decoder
The model consists of two distinct neural network components connected via a bottleneck representation:

![Encoder-Decoder Seq2Seq Architecture](outputs/encoder-decoder.png)

1.  **The Encoder**:
    *   Takes the source sequence word-by-word.
    *   Maintains a recurrent hidden state vector that updates at each time step.
    *   At the final word, the Encoder compresses the entire semantic meaning of the sentence into its final hidden state $h_T$. This vector is called the **Context Vector** (or the semantic bottleneck).
2.  **The Context Vector**:
    *   Acts as a "summary" or translation pivot of the input sentence.
    *   Passed directly as the initial hidden state to the Decoder.
3.  **The Decoder**:
    *   Initialized with the context vector.
    *   Generates the target sequence token-by-token.
    *   At each step, the Decoder uses its own previous output word and updated hidden state to predict the next word in the target language.

### Step-by-Step Translation Walkthrough
Let's trace how the sentence *"Je suis fatigué"* is translated into *"I am tired"* using a Seq2Seq model:

1.  **Encoding Phase**:
    - Step 1: Input `"Je"` enters the Encoder. The hidden state updates to $h_1$ (capturing *"Je"*).
    - Step 2: Input `"suis"` enters. Hidden state updates to $h_2$ (capturing *"Je suis"*).
    - Step 3: Input `"fatigué"` enters. Hidden state updates to $h_3$ (the final **Context Vector** representing the full semantic meaning).
2.  **Decoding Phase**:
    - The Encoder hands the context vector $h_3$ over to the Decoder as its initial hidden state $s_0$.
    - Step 1: The Decoder is fed a special start token `<SOS>` (Start of Sentence). Using $s_0$ and `<SOS>`, it predicts `"I"` and updates its state to $s_1$.
    - Step 2: The predicted word `"I"` is fed as input. Using $s_1$ and `"I"`, the Decoder predicts `"am"` and updates its state to $s_2$.
    - Step 3: The predicted word `"am"` is fed as input. The Decoder predicts `"tired"` and updates its state to $s_3$.
    - Step 4: The predicted word `"tired"` is fed. The Decoder predicts the end token `<EOS>` (End of Sentence), signaling that translation is complete!

### Sequence-to-Sequence Modeling Advantages & Limitations
*   **Advantages**:
    - Maps inputs of any length to outputs of any length.
    - Captures long-range semantics and word ordering context before beginning generation.
*   **Limitations (The Information Bottleneck)**:
    - Attempting to compress a very long sentence (e.g. 50 words) into a single fixed-size context vector vector results in massive information loss. The model "forgets" the beginning of long sentences—this limitation directly motivated the emergence of the **Attention Mechanism**.
